```python
import cv2
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
img = np.zeros((400, 400, 3), dtype = "uint8")
cv2_imshow(img)
```

```python
cv2.line(img,(20,160),(200,200),(12,168,200),10)
cv2_imshow(img)
```

```python
img = np.zeros((400, 400, 3), dtype = "uint8")
cv2.rectangle(img,(100,100),(300,300),(255, 0, 205),10)
cv2_imshow(img)
```

```python
img = np.zeros((400, 400, 3), dtype = "uint8")
pts = np.array([[50,50],[30,100],[100,130],[120,150],[190,200]],np.int32)
cv2.polylines(img,[pts],True,(105,0,255))
cv2_imshow(img)
```
