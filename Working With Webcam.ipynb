## 1. Working With Webcam

- <code>cv2.VideoCapture()</code> : This is used for accessing the Webcam feature. 
   1. <i>0 :</i> For inbuilt webcam on laptop
   2. <i>(1) or (2) or .... :</i> For external webcams attached to laptop
   
<code>cam = cv2.VideoCapture(0)</code> (Here, camera is a file descriptor)

<h4>For Releasing the camera resources :</h4></br>
<code>cam.release()</code>

- <code>cv.waitKey(0)</code> waits indefinitely for a key press before proceeding, which effectively pauses your loop after displaying the first frame. To continuously capture and display frames from your webcam, you should use a small positive integer as the argument to <code>cv.waitKey()</code>, such as <code>1</code>, which waits for 1 millisecond.

- When you will press enter or any other key, it will execute the consecutive frames.

- We need to introduce a stopping condition also. That if we press escape, then it will break from the loop.

- If <code>cv.waitKey(30) :</code> Introduces <code>(1000//30) = 33.333 = 33</code>frames per second. </br></br><code>cv.destroyAllWindows() :</code> Destroys or closes the window frames.

```python
cam = cv.VideoCapture(0)
while True:

    _, img = cam.read()
    cv.imshow("Frame",img)
    key = cv.waitKey(1) 
    
    if key == 27:
        cam.release()
        break
cv.destroyAllWindows()
```
